```python
import pandas as pd
```

```python
movies = pd.read_csv('../data-repo/imdb_1000.csv')

movies.count()
```

```python
# old way
star_filter = []
for rate in movies['star_rating']:
    if rate > 8.5:
        star_filter.append(True)
    else:
        star_filter.append(False)

# 转换为Series
star_filter_series = pd.Series(star_filter)
star_filter_series.head()

# 返回过滤后数据
filtered_movies = movies[star_filter_series]

filtered_movies
```

```python
# pandas way

pd_fiters = movies['star_rating'] >= 8.5

movies[pd_fiters]
movies.loc[pd_fiters, 'title']
```
